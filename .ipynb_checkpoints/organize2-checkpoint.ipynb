```python
#Organizing the data
import json
class organize():
    def __init__(self,fullpath):
        with open(fullpath) as f:
            jsondata = f.read()
        py_obj=json.loads(jsondata)
        self.data=py_obj["data"]   #list of the original data 
        self.country_list=self.__country_list()[0]#list of all the countries in the data
        self.country_region=self.__country_region_dic()[0] #dic of all the regions,{countries:[region1,region2...],...}
        self.country_city=self.__country_city_dic() #dic of all the cities,{countries:[city1,city2...],...}
        self.number_city=self.__count("self.country_city")#number of cities by country
        self.number_region=self.__count("self.country_region")#number of ip address by region
        self.number_ip=self.__count_ip()#number of ip address by country
        self.coordinate=self.__country_list()[1]
        self.coordinate=self.__country_region_dic()[1]
   
    def __country_list(self):
        country_list=[]
        country_cor=[]
        for dic in self.data:
            if len(dic["countryName"])==0:
                continue
            if dic["countryName"] in country_list:
                continue
            else:
                country_list.append(dic["countryName"])
                country_cor.append([dic["countryName"],(dic["latitude"],dic["longitude"])])
        return country_list,country_cor
#   print("Numbers of the countries included in the data: "+str(len(country_list)))
    def __country_city_dic(self):
        countrycity_dic={}
        country_list=self.country_list
        for c_name in country_list:
            state=0
            for dic in self.data:
                if state==0:
                    countrycity_dic[c_name]=[]
                    state=1
                if c_name==dic["countryName"] and not(dic["cityName"] in countrycity_dic[c_name]) and dic["cityName"] != "-":
                    countrycity_dic[c_name].append(dic["cityName"])
        return countrycity_dic
    def __country_region_dic(self):
        countryregion_dic={}
        region_cor=[]
        country_list=self.country_list
        for c_name in country_list:
            state=0
            for dic in self.data:
                if state==0:
                    countryregion_dic[c_name]=[]
                    state=1
                if c_name==dic["countryName"] and not(dic["regionName"] in countryregion_dic[c_name]) and dic["regionName"] != "-":
                    countryregion_dic[c_name].append(dic["regionName"])
                    region_cor.append([dic["regionName"],(dic["latitude"],dic["longitude"])])
                    
        return countryregion_dic,region_cor

        # pass on dictionary that corresponds country to city and return the number of cities in each country
    def __count(self,li):
        num = []
        country_city=self.country_city
        country_region=self.country_region
        if li=="self.country_city":
            for c_name in self.country_list:
                num.append([c_name, len(country_city[c_name])])
        else:    
            for c_name in self.country_list:
                num.append([c_name, len(country_region[c_name])])
        return num

    def __count_ip(self):
        num_ip=[]
        data=self.data[:]
        for con in self.country_list:
            c=0
            for dic in data:
                if dic["countryName"]==con:
                    c+=1
            num_ip.append([con,c])

        return num_ip

    #below are the functions that can be used
    def summary(self):
        print("The total number of ip address: "+str(len(self.data)))
        print("The total number of the countries included in the data: "+str(len(self.country_list)))
        city=0
        region=0
        for i in self.number_city: city+=i[1]
        for i in self.number_region: region+=i[1]
        print("The total number of the regions included in the data: "+str(region))
        print("The total number of the cities included in the data: "+str(city))
        
    def check_ip(self,countryName):
        ip_list=list(filter(lambda dic:dic["countryName"]==countryName,self.data))
        return [x["ipAddress"] for x in ip_list]
    
    #count the number of ip address of every city in the data.
    def count_ip_by_city(self):
        num_ip_city = []             
        for con in self.country_list:
            for city in self.country_city[con]:
                c = 0
                for dic in self.data:
                    if dic["cityName"] == city:
                        c += 1
                num_ip_city.append([city, c])
                # num_ip_city.append([city,len(list(filter(lambda x:x["cityName"]==city,self.data)))])
        return num_ip_city
    #count the number of ip address of the region in the data. it takes minutes due to the huge size of the data
    def count_ip_by_region(self):
        num_ip_region = []             
        for con in self.country_list:
            for region in self.country_region[con]:
                c = 0
                for dic in self.data:
                    if dic["regionName"] == region:
                        c += 1
                num_ip_region.append([region, c])
                # num_ip_city.append([city,len(list(filter(lambda x:x["cityName"]==city,self.data)))])
        return num_ip_region
    #pass in list with [[string,int],[string,int]...] format and return the sorted version
    def sort(self,lis):
        return sorted(lis,key=lambda contents:contents[1],reverse=True)
```
