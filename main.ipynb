```python
"""The geolocation API"pyipinfodb"gives access to the 
    database of an IP address. if the address is passed in, 
    it gives back the geolocation(city, region, country Name,
    latitude and longitude...) in a dictionay format """
from pyipinfodb import pyipinfodb

ip_if=pyipinfodb.IPInfo("c8825bed76ebd5569eea87a46a0e103b7f6de30335343702aadd9d4ebba95949")
print(ip_if.get_city('13.102.40.0'))
```

```python
"""the organize object organizes the JSON ip info data saved in the ip.json file and 
provides methods to analyze the raw data(filtering,extracting specified info) 
This json file was created by running ipAnalyze/collect.py, 
which relies on the api shown above. """
from ipAnalyze import organize
```

```python
import os
path_to_data = os.getcwd()+"/ip.json"
organizeData = organize.organizeIpInfo(path_to_data)
```

```python
#summary of the data
organizeData.summary()
```

```python
"filtering raw data"
organizeData.extract_info('cityName','Tokyo')
```

```python
#countries in the data
organizeData.country
```

```python
#returns the regions/cities of the country passed in
print(organizeData.extract_region("United Kingdom"))
print()
print(organizeData.extract_city("United Kingdom")[:80])
```

```python
#Returns the Ip addresses of a passed country/region/city
organizeData.check_ip("Tokyo","city")
organizeData.check_ip("England","region")
organizeData.check_ip("Australia","country")
```

```python
#returns the coordinates of countries/regions
organizeData.region_coordinate
organizeData.country_coordinate
```

```python
#returns pandas Data frame by country/region/city 
countryDf,regionDf,cityDf = organizeData.pandasDF()
```

```python
countryDf.head(10)
```

```python
from ipAnalyze import visualize as vs
"""Visualizing the data using seaborn
 Pass in the dataframe and the index/name you are looking fo"""
```

```python
#pie chart: displays the percentage of the numOfIp/region/city by country 
#and numOfIp by region/city
vs.pie(dtype='numberOfIp',Df=countryDf)
vs.pie(index=[1,5],dtype='numberOfCity',Df=countryDf)
vs.pie(index=[1,5,7,2],dtype='numberOfIp',Df=regionDf)
```

```python
#Overview of the data
#Displaying the top 1~20th countries
vs.barplot_Overview(index=[1,10],Df=countryDf)
```

```python
#Displaying the number of ip addresses of the region of the index that was passed in 
from ipAnalyze import visualize as vs
vs.barplot_Ip(index=[3,2,1,4,10,11,5,8,12,9,6,7,43],Df=regionDf)
```

```python
#Displaying the number of IP addresses of passed in cities in the UK
print(organizeData.extract_city("United Kingdom")[:30])
vs.barplot_Ip(name=['London_United Kingdom', 'Leeds_United Kingdom', 'Liverpool_United Kingdom', 'Birmingham_United Kingdom',
                    'Manchester_United Kingdom','Middlesbrough_United Kingdom','York_United Kingdom','Chard_United Kingdom','Newport_United Kingdom'],
              Df=cityDf)
```

```python
#Displaying the distribution of Ip by country/region/city
vs.hist(start=10,end=15000,Df=countryDf)
vs.hist(start=3000,end=20000,Df=regionDf)
vs.hist(start=20,end=500,Df=cityDf)
```
